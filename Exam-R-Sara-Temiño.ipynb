#### Task 1: Load the files GDP.csv and HDI.csv using read.csv()

```r
GDP = read.csv(file='GDP.csv')
HDI = read.csv(file='HDI.csv')
```

#### Task 3: Glue together the two data frames by columns (one next to the other) using cbind(). We want the province to match on the same rows, so you need first to sort the rows of the two data frames by Province, using order() 

```r
df1 = GDP[order(GDP$Province), ]
df2 = HDI[order(HDI$Province), ]
df3 = cbind(df1, df2)
```

#### Task 4: The new data frame will contain two columns names "Province". remove one of the columns

```r
df4 = df3[,-4]
```

#### Task 5: Check that everything is fine with head() and tail(). The final data frame should have 4 columns: Province GDP.per.capita, CCAA and HDI. With each province matching its CCAA. (Fig, 1)

```r
head (df4)
tail (df4)
```

#### Task 6: Create scatterplot of text using geom_point(). Plot the GDP.per.capita (x-axis) against the HDI (y-axis). Color the points by CCAA (use color=CCAA inside the aes() function of geom_point).

```r
library(ggplot2)
theme_set(theme_bw())
options(repr.plot.width = 20, repr.plot.height = 10, repr.plot.res = 100)
ggplot(df4, aes(x = GDP.per.capita, y = HDI, color = CCAA)) + geom_point(size=4) + theme(text = element_text(family = "Arial", size = 20))
```

#### Task 7: Do the same but use geom_text(). Specify as labels the names of the provinces, using aes(label=Province) inside geom_text. Color by CCAA. Also specify check_overlap=T so that the names do not overlap, if you want. (Fig, 2)

```r
theme_set(theme_bw())
options(repr.plot.width = 20, repr.plot.height = 10, repr.plot.res = 100)
ggplot(df4, aes(x = GDP.per.capita, y = HDI, color = CCAA)) + geom_text(data=df4, aes(label=Province), size=6, check_overlap=T) + theme(text = element_text(family = "Arial", size = 20))
```

#### Task 8: Calculate the mean and standard deviation of GDP and HDI, by CCAA, using aggregate ().
#### Task 9: Create a new data frame where you put each CCAA with the corresponding mean GDP, mean HDI, stdev GDP and stdev HDI

```r
#Task8
meansbyCCAA = aggregate(cbind(GDP_m = GDP.per.capita, HDI_m = HDI) ~ CCAA, FUN = mean, data=df4)
sdbyCCAA = aggregate(cbind(GDP_sd = GDP.per.capita, HDI_sd = HDI) ~ CCAA, FUN = sd, data=df4)
#Task9
df5 = cbind(meansbyCCAA, sdbyCCAA)[-4]
```

#### Task 10: Madrid is a single province CCAA, so you will see that its standar deviation is NA. Replace those NAs with zeros, using is.na(). (Fig, 3)

```r
df5[,4:5][is.na(df5[,4:5])] <- 0
df5
```

#### Task 13: Add also vertical and horizontal error bars to express the variablity for each CCAA (with geom_errorbar and geom_errorbarh). For geom_errorbar() you have to specify the aestetic paramters x, y, ymin and ymax. For geom_errorbarh you have to specify the aestetic paramters x, y, xmin and xmax.
